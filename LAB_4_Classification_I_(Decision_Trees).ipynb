```python
import os
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
from graphviz import Source
from sklearn.datasets import load_iris, load_wine, load_breast_cancer
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Iris dataset

```python
data = load_iris()
X = data.data[:, 2:]
y = data.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
tree_clf.fit(X_train,y_train)
```

```python
PROJECT_ROOT_DIR = '.'
CHAPTER_ID = "decision_trees"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)
```

```python
export_graphviz(
        tree_clf,
        out_file=os.path.join(IMAGES_PATH, "iris_tree.dot"),
        feature_names=data.feature_names[2:],
        class_names=data.target_names,
        rounded=True,
        filled=True
    )

Source.from_file(os.path.join(IMAGES_PATH, "iris_tree.dot"))
```

```python
data.feature_names
```

```python
iris_data = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
yhat = tree_clf.predict(X_test)
accuracy_score(y_test, yhat)
```

# Wine dataset

```python
data = load_wine()
X = data.data[:, 2:]
y = data.target
tree_clf = DecisionTreeClassifier(max_depth=2)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
tree_clf.fit(X_train,y_train)
data.feature_names[2:]
PROJECT_ROOT_DIR = '.'
CHAPTER_ID = "decision_trees"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)
export_graphviz(
        tree_clf,
        out_file=os.path.join(IMAGES_PATH, "iris_tree.dot"),
        feature_names=data.feature_names[2:],
        class_names=data.target_names,
        rounded=True,
        filled=True
    )

Source.from_file(os.path.join(IMAGES_PATH, "iris_tree.dot"))
```

```python
wine_data = pd.DataFrame(load_wine().data)
```

```python
sns.heatmap(wine_data.corr())
```

```python
sns.displot(wine_data[[0,1,2,3]], kind='kde')
```

```python
sns.boxenplot(data=wine_data.drop(columns=[4,12]))
```

```python
yhat = tree_clf.predict(X_test)
accuracy_score(y_test, yhat)
```

```python
data = load_breast_cancer()
X = data.data[:, 2:]
y = data.target
tree_clf = DecisionTreeClassifier(max_depth=2)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
tree_clf.fit(X_train,y_train)
data.feature_names[2:]
PROJECT_ROOT_DIR = '.'
CHAPTER_ID = "decision_trees"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)
export_graphviz(
        tree_clf,
        out_file=os.path.join(IMAGES_PATH, "iris_tree.dot"),
        feature_names=data.feature_names[2:],
        class_names=data.target_names,
        rounded=True,
        filled=True
    )

Source.from_file(os.path.join(IMAGES_PATH, "iris_tree.dot"))
```
